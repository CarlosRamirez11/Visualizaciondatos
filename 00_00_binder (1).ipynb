## loc y iloc
[row_inicial:row_final, col_inicial: col_final]
Si usa loc, usa nombres de los renglones y nombres de las columnas

SI usa iloc, usa indices de los renglones y las columnas

```python
df.loc[2:7, 'Age':'Height']
```

```python
df.iloc[2:10, 3:6]
```

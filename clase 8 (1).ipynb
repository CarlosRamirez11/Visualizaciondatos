```python
x = df['Height']
y = df['Weight']
x= x.to_frame()
y= y.to_frame()
type(y)
```

```python
#pip install statsmodels
```

```python
import statsmodels.api as sm
```

```python
#pip install sklearn
```

```python
from sklearn import linear_model
```

```python
lm = linear_model.LinearRegression()
lm
```

```python
model_lm = lm.fit(y,x)
```

```python
print('El intercepto es \t:', model_lm.intercept_)
print('El coeficiente es \t:', model_lm.coef_)
print('El valor de R2 es \t:', model_lm.score(y,x))
```

```python
model_ols = sm.OLS.from_formula('x~y', data = df) #Height = b0 + b1*Weight; //// Weight = b0 + b1*Height
result= model_ols.fit()
result.summary()
```

```python
vars = ['SMQ020', 'RIAGENDR', 'RIDAGEYR', 'BMXWT', 'BMXHT', 'BMXBMI', 'BMXWAIST', 'HIQ210']
```

```python
df = df[vars].dropna() #dropna es para eliminar los datos NaN
df
```

```python
df.corr()
```

## BMXWT = b0 + b1*RIDAGEYR

```python
model= sm.OLS.from_formula('BMXWT~RIDAGEYR', data =df)
result= model.fit()
result.summary()
```

```python
df['RiaGENDR'] = df.RIAGENDR.replace({1: 'Male', 2: 'Female'})
df
```

### BMXWT = b0 + b1 RIDAGEYR + b2RiaGENDR

### BMXWT = b0 + b1*RIDAGEYR + b2 (Male + Female)

### Female
